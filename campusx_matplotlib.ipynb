**3D Surface plots**

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
```

```python
# mesh grid : forms a grid
xx,yy = np.meshgrid(x,y)
```

```python
xx.shape
```

```python
yy.shape
```

```python
z = xx**2 + yy**2
z.shape
```

```python
fig = plt.figure(figsize=(12,8))
ax = plt.subplot(projection = '3d')
p = ax.plot_surface(xx,yy,z,cmap = 'viridis')
fig.colorbar(p)
```

```python
z = np.sin(xx)+ np.cos(yy)
fig = plt.figure(figsize=(12,8))
ax = plt.subplot(projection = '3d')
p = ax.plot_surface(xx,yy,z,cmap = 'viridis')
fig.colorbar(p)
```

**contour plot 3D to 2D**

```python
# top view of 3D surface plot
z = xx**2 + yy**2
fig = plt.figure(figsize=(12,8))
ax = plt.subplot()
p = ax.contour(xx,yy,z,cmap = 'viridis')
fig.colorbar(p)
```

```python
# top view of 3D surface plot
z = xx**2 + yy**2
fig = plt.figure(figsize=(12,8))
ax = plt.subplot()
p = ax.contourf(xx,yy,z,cmap = 'viridis')
fig.colorbar(p)
```
